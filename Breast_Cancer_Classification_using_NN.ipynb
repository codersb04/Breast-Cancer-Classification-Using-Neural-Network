## **Import Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

## **Data Collection and Processing**

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
data_frame['Target'] = breast_cancer_data.target
```

Splitting the features and Target

```python
X = data_frame.drop('Target', axis=1)
Y = data_frame['Target']
```

```python
print(X)
Y
```

Splitting the data frame into Training and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
X_test_std
```

## **Building Neural Network**

```python
# Importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
from keras.src.layers.serialization import activation
# Setting up the layers of Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
from keras.src.optimizers import Optimizer
# Compiling the Neural Network

model.compile(optimizer= 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# Training the model

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

## Visualizing accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('accuray')
plt.xlabel('epoch')

plt.legend(['training_data', 'validation_data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loass')
plt.xlabel('epoch')

plt.legend(['training_data', 'validation_data'], loc = 'upper right')
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print("Accuracy of test data: ", round(accuracy*100,2))
```
